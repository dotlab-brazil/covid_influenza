# A célula abaixo elimina as colunas com as variáveis indesejadas.

```python
plan = pd.read_excel('/content/drive/MyDrive/mestrado - covid 2023/Variáveis indesejadas.xlsx')
df10 = df.drop(columns = plan.Siglas)
df10.head()
```

```python
## Verificando a quantidade de dados nulos
df10.isnull().sum()
```

```python
# Esta célula converte os valores de data que estão no formato string/object para o 
# formato datetime
df10.dt_evoluca = pd.to_datetime(df10.dt_evoluca)
df10.dt_encerra = pd.to_datetime(df10.dt_encerra)
df10.dt_notific = pd.to_datetime(df10.dt_notific)
```

```python
df10.dtypes
```

```python
# Cria duas novas variáveis
df10['DIAS_ATE_EVOLUCAO']= abs(df10.dt_evoluca - df10.dt_notific)
df10['DIAS_ATE_ENCERRAMENTO'] = abs(df10.dt_encerra - df10.dt_notific)

## Substitui os valores faltantes por 0 days
df10.dt_evoluca = df10.dt_evoluca.fillna(0)
df10.dt_notific = df10.dt_notific.fillna(0)
df10.dt_encerra = df10.dt_encerra.fillna(0)
df10.DIAS_ATE_EVOLUCAO = df10.DIAS_ATE_EVOLUCAO.fillna(pd.Timedelta(days = 0))
df10.DIAS_ATE_ENCERRAMENTO = df10.DIAS_ATE_ENCERRAMENTO.fillna(pd.Timedelta(days = 0))
```

```python
x,y = 'classi_fin', 'evolucao'

df1 = df30.groupby(x)[y].value_counts(normalize=True)
df1 = df1.mul(100)
df1 = df1.rename('percent').reset_index()
sns.set(font_scale=2)

g = sns.catplot(x=x,y='percent',hue=y,kind='bar',data=df1, size=10)
g.ax.set_ylim(0,100)

for p in g.ax.patches:
    txt = str(p.get_height().round(2)) + '%'
    txt_x = p.get_x() 
    txt_y = p.get_height()
    g.ax.text(txt_x,txt_y,txt)
```
